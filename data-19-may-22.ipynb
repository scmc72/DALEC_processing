```python
# let's plot the spectra of a specific sample:

spect_wavelengths = dalecLoad.load_DALEC_spect_wavelengths('data/LOG_0046.dtf')

sample_10 = log46.loc[10, :]
Lu_10 = dalecLoad.uniform_grid_spectra(sample_10, spect_wavelengths, param='Lu')
Lsky_10 = dalecLoad.uniform_grid_spectra(sample_10, spect_wavelengths, param='Lsky')
Ed_10 = dalecLoad.uniform_grid_spectra(sample_10, spect_wavelengths, param='Ed')

# very simple Rrs calculation
RHO = 0.028
Rrs_10 = (Lu_10[:, 1] - (RHO * Lsky_10[:, 1])) / Ed_10[:, 1]
```

```python
fig, ax = plt.subplots(1, 1, figsize=[10, 6])

ax.plot(Lu_10[:, 0], Rrs_10)

ax.plot([620, 620], [0.02, -0.00], '--', label='Phycocyanin (620nm)')
ax.plot([680, 680], [0.02, -0.00], '--', label='Chl-a (680nm)')
ax.plot([440, 440], [0.02, -0.00], '--', label='Chl-a (440nm)')

plt.rc('axes', labelsize=14) #fontsize of the x and y labels

ax.set_xlabel('Wavelength (nm)')
ax.set_ylabel('$R_{rs} (sr^{-1}$)')
ax.grid()

plt.rc('legend', fontsize=14) #fontsize of the legend
plt.legend()
plt.show()
```

```python
# let's plot the inverse of remote sensing reflectance!

fig, ax = plt.subplots(1, 1, figsize=[12, 6])

ax.plot(Lu_10[:, 0], 1/Rrs_10)


ax.set_xlabel('Wavelength (nm)')
ax.set_ylabel('$1 / R_{rs} (sr$)')
ax.grid()

ax.set_ylim(0, 250)
ax.set_xlim(400, 900)

plt.show()
```
